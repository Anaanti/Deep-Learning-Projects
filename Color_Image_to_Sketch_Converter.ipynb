```python
from scipy import ndimage
import imageio.v3 as iio
import numpy as np
import matplotlib.pyplot as plt
```

# Load the Data

```python
img = iio.imread('imageio:chelsea.png')
plt.imshow(img)
plt.show()
```

# Converting to Grayscale

```python
gray = np.dot(img[..., :3],[0.299, 0.587, 0.114])
plt.imshow(gray, cmap='gray')
plt.show()
```

# Invert the grayscale image

```python
inverted = 255 - gray
plt.imshow(inverted)
plt.show()
```

# Gausian Blur

```python
blurred = ndimage.gaussian_filter(inverted, sigma=6)
plt.imshow(blurred)
plt.show()
```

# Inverting the blurred image

```python
inverted_blurred = 255 - blurred
plt.imshow(inverted_blurred)
plt.show()
```

# Creating Pencil Sketch

```python
sketch = gray / inverted_blurred
sketch = sketch * 255
sketch = np.clip(sketch , 0, 255).astype(np.uint8)
plt.imshow(sketch, cmap='gray')
plt.axis('off')
plt.show()
```
